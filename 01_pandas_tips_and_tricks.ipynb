# 16- Splitting a string into multiple columns

```python
# Splitting a string into multiple columns
df = pd.DataFrame({'Name' : ['Adnan Hyder', 'Farhan Ali', 'Ameer Hamza'],
                   'Location' : ['Jacobabad, Pakistan', 'Nawabshah, Pakistan', 'Karachi, Pakistan']})
df
```

```python
# Separate the Name column by space and create two new columns

df[['First Name', 'Last Name']] = df['Name'].str.split(' ', expand=True)
df[['City', 'Country']] = df['Location'].str.split(',', expand=True)
```

```python
df
```

```python
# Refine manipulated data
df = df[['First Name', 'Last Name', 'City', 'Country']]
df
```

# 23- Reshaping a Dataframe

```python
distance = pd.DataFrame([['12345', 100, 200, 300], ['34567', 400, 500, 600], ['67890', 700, 800, 900]],
                        columns=['zip', 'factory', 'warehouse', 'retail'])
distance.head()
```

```python
sns.barplot(x='zip', y='factory', data=distance)
```

```python
distance_long = distance.melt(id_vars="zip", var_name = 'location_type', value_name='distance')
```

```python
sns.barplot(x='zip', y='distance', hue='location_type', data=distance_long)
```
